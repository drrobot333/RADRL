```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('C:/Users/Suhwan/Desktop/suhwan/connection_test/paper_figure/experiment/complete num/node5_lowband_compnum.csv')  # Replace 'your_file.csv' with the actual file path

maximum = data.max(numeric_only=True).max()

x = data['jobrate']

# maximum = 30


# Combine the data into a list
data = [data['our']/maximum, data['CDCR']/maximum, data['WSPT']/maximum]
fig, ax = plt.subplots()

boxplot = ax.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue'), showfliers=False, medianprops={"linewidth":3},)

median_line = boxplot['medians'][0]
x_median = median_line.get_xdata()[1]
y_median = median_line.get_ydata()[0]
print(x_median, y_median)

# Find the coordinates of the minimum and maximum lines
minimum_line = boxplot['whiskers'][0]
x_min = minimum_line.get_xdata()[1]
y_min = minimum_line.get_ydata()[1]  # The lower end of the whisker
print(x_min, y_min)

maximum_line = boxplot['whiskers'][1]
x_max = maximum_line.get_xdata()[1]
y_max = maximum_line.get_ydata()[1]  # The lower end of the whisker

path = boxplot['boxes'][0].get_path().vertices
print(path)
x_q1 = (path[0][0] + path[1][0]) / 2
y_q1 = path[0][1]
print(x_q1, y_q1)

x_q3 = (path[2][0] + path[3][0]) / 2
y_q3 = path[2][1]
print(x_q3, y_q3)


# Add annotations to indicate the range

ax.annotate('Median', xy=(x_q1 + 0.14, y_median), xytext=(x_median + 0.2, y_median),
            arrowprops=dict(arrowstyle='->'), ha='left', va='center')
ax.annotate('Min', xy=(x_min + 0.07, y_min), xytext=(x_min + 0.35, y_min),
            arrowprops=dict(arrowstyle='->'), ha='left', va='center')
ax.annotate('Max', xy=(x_max + 0.07, y_max), xytext=(x_max + 0.35, y_max),
            arrowprops=dict(arrowstyle='->'), ha='left', va='center')
ax.annotate('75%', xy=(x_q1 + 0.14, y_q3), xytext=(x_q1 + 0.35, y_q3),
            arrowprops=dict(arrowstyle='->'), ha='left', va='center')
ax.annotate('25%', xy=(x_q1 + 0.14, y_q1), xytext=(x_q1 + 0.35, y_q1),
            arrowprops=dict(arrowstyle='->'), ha='left', va='center')


# Customize the colors of the boxes
colors = ['blue', 'orange', 'green']
for patch, color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)

# Add labels and a legend
plt.xlabel('')
plt.ylabel('Number of Complete DNN Jobs', fontsize=13)

# plt.rc('xtick', labelsize=11)  # x    
# plt.rc('ytick', labelsize=15)  # y   

# Specify the labels for each boxplot
legend_handles = [Patch(facecolor=color) for color in colors]
legend_labels = ['ours', 'CDCR', 'WSPT']
plt.legend(legend_handles, legend_labels, loc='lower left', fontsize=10)

plt.ylim(0.6, 1.05)
plt.xticks(range(1, len(legend_labels) + 1), legend_labels)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a pandas DataFrame
file_name = "reward_convergence"
data_5 = pd.read_csv(f'C:/Users/Suhwan/Desktop/suhwan/connection_test/paper_figure/experiment/converge/node5.csv')  # Replace 'your_file.csv' with the actual file path

data_10 = pd.read_csv(f'C:/Users/Suhwan/Desktop/suhwan/connection_test/paper_figure/experiment/converge/node10.csv')  # Replace 'your_file.csv' with the actual file path

data_length_10 = 700
data_length_5 = 700



x_10 = data_10['Step'][:data_length_10]
x_5 = data_5['Step'][:data_length_5]

data_10 = data_10.iloc[:, :]['Value']
data_5 = data_5.iloc[:, :]['Value']

window_size = 50

# smooth data
padded_data_10 = np.pad(data_10, (window_size // 2, window_size // 2), mode='edge')
padded_data_5 = np.pad(data_5, (window_size // 2, window_size // 2), mode='edge')


# Apply the moving average
smoothed_data_5 = np.convolve(padded_data_5, np.ones(window_size) / window_size, mode='valid')[:data_length_5]
smoothed_data_10 = np.convolve(padded_data_10, np.ones(window_size) / window_size, mode='valid')[:data_length_10]

print(len(smoothed_data_10))
print(len(smoothed_data_5))

maximum_10 = smoothed_data_10.max()
maximum_5 = smoothed_data_5.max()

print(maximum_10)
print(maximum_5)

smoothed_data_5 /= maximum_5

smoothed_data_10 /= maximum_10

# Create the plot
plt.plot(x_10, smoothed_data_10, label='10 nodes', linestyle="solid")
plt.plot(x_10, smoothed_data_5, label='5 nodes', linestyle="dashed")
# plt.plot(x, data['CDCR'], label='CDCR', marker='x', markersize=10)
# plt.plot(x, data['WSPT'], label='WSPT', marker='^', markersize=10)

plt.tick_params(axis='x', labelsize=12)
plt.tick_params(axis='y', labelsize=12)

# Add labels and a legend
plt.xlabel('Episode', fontsize=13)
plt.ylabel('Normalized Reward', fontsize=13)
# plt.ylim(0.0, 1.2)
plt.xlim(-300, 7800)

plt.legend(loc='lower right', fontsize=15)

# Display the plot

plt.savefig(f'C:/Users/Suhwan/Desktop/suhwan/connection_test/paper_figure/experiment/converge/{file_name}.png')
```
